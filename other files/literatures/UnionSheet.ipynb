```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
xls = pd.ExcelFile("2025中科院分区表-分领域整理.xlsx")
sheet_names = xls.sheet_names

dfs = pd.read_excel("./2025中科院分区表-分领域整理.xlsx", sheet_name=sheet_names)
df0 = dfs["巨型期刊"]

new_header = df0.iloc[0]
df0 = df0[1:]
df0.columns = new_header

df0
```

```python
total_df = df0
columns = ["期刊名称", "ISSN", "2025大类分区"]
for k in sheet_names[1:]:
    v = dfs[k]
    v.columns = columns
    pd.DataFrame.insert(v, 2, column="大类", value=k)
    total_df = pd.concat([total_df, v], axis=0, ignore_index=True)

total_df
```

```python
total_df.to_excel("total_excel.xlsx")
```
